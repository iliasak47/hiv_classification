The total molecular weight — it gives a rough idea of the molecule’s size, which affects how well it can move through the body or reach its target.

We have to normalize our data, we will do this in a python script

```python
sns.histplot(df_processed,x='TPSA')
plt.show()
```

```python
df_processed['TPSA'].describe()
```

The topological polar surface area — it reflects how well the molecule can form hydrogen bonds, which impacts solubility and biological activity

We will also normalize this feature

```python
sns.histplot(df_processed,x='NumRotatableBonds')
plt.show()
```

```python
df_processed["NumRotatableBonds"].describe()
```

The number of freely rotating single bonds — it measures flexibility, which can affect how well the molecule fits with a target.

```python
sns.histplot(df_processed,x="NumHDonors")
plt.show()
```

```python
df_processed['NumHDonors'].describe()
```

The number of hydrogen bond donors — atoms that can "give" a hydrogen to interact with proteins or water.

```python
sns.histplot(df_processed,x="NumHAcceptors")
plt.show()
```

```python
df_processed["NumHAcceptors"].describe()
```

The number of hydrogen bond acceptors — atoms that can "receive" a hydrogen; also important for interactions and solubility.

```python
sns.histplot(df_processed,x="NumAromaticRings")
plt.show()
```

```python
df_processed['NumAromaticRings'].describe()
```

The number of aromatic rings — flat, stable ring structures that help molecules interact with biological targets (common in drugs).

```python
sns.histplot(df_processed,x="LogP") 
plt.show()   
```

```python
df_processed['LogP'].describe()
```

A measure of how lipophilic (fat-loving) or hydrophilic (water-loving) a molecule is — this impacts absorption and bioavailability.

Let's see how these features correlate

```python
mat_corr=df_processed[["MolWt", "TPSA", "LogP", "NumRotatableBonds","NumHDonors","NumHAcceptors","NumAromaticRings","LogP","standard_value"]].corr()
sns.heatmap(mat_corr,annot=True,cmap="Blues")
plt.show()
```

Low correlations between individual features and IC50 values are expected, as biological activity is driven by complex, nonlinear interactions rather than simple linear relationships.

Let's create the active feature 

```python
df_processed_clean=df_processed[df_processed['standard_relation']=="="]
```

```python
df_processed_clean['standard_value'] = pd.to_numeric(df_processed_clean['standard_value'], errors='coerce')
```

```python
df_processed_clean['active'] = (df_processed_clean['standard_value'] < 1000).astype(int)
```

```python
df_processed_clean['active'].value_counts()
```

We have a target feature with 58% of active and 42% of inactive wich is pretty equilibrated

```python
features = ["MolWt", "TPSA", "LogP", "NumRotatableBonds", "NumHDonors", "NumHAcceptors", "NumAromaticRings"]

for feature in features:
    plt.figure(figsize=(6, 4))
    sns.boxplot(data=df_processed_clean, x="active", y=feature)
    plt.title(f"{feature} by Activity")
    plt.xlabel("Active (1 = active, 0 = inactive)")
    plt.ylabel(feature)
    plt.tight_layout()
    plt.show()
```
