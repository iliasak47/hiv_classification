A measure of how lipophilic (fat-loving) or hydrophilic (water-loving) a molecule is â€” this impacts absorption and bioavailability.

Let's see how these features correlate

```python
mat_corr=df_processed[["MolWt", "TPSA", "LogP", "NumRotatableBonds","NumHDonors","NumHAcceptors","NumAromaticRings","LogP","standard_value"]].corr()
sns.heatmap(mat_corr,annot=True,cmap="Blues")
plt.show()
```

Low correlations between individual features and IC50 values are expected, as biological activity is driven by complex, nonlinear interactions rather than simple linear relationships.

Let's create the active feature 

```python
df_processed_clean=df_processed[df_processed['standard_relation']=="="]
```

```python
df_processed_clean['standard_value'] = pd.to_numeric(df_processed_clean['standard_value'], errors='coerce')
```

```python
df_processed_clean['active'] = (df_processed_clean['standard_value'] < 1000).astype(int)
```

```python
df_processed_clean['active'].value_counts()
```

We have a target feature with 58% of active and 42% of inactive wich is pretty equilibrated

```python
features = ["MolWt", "TPSA", "LogP", "NumRotatableBonds", "NumHDonors", "NumHAcceptors", "NumAromaticRings"]

for feature in features:
    plt.figure(figsize=(6, 4))
    sns.boxplot(data=df_processed_clean, x="active", y=feature)
    plt.title(f"{feature} by Activity")
    plt.xlabel("Active (1 = active, 0 = inactive)")
    plt.ylabel(feature)
    plt.tight_layout()
    plt.show()
```
